```python
import cv2
```

```python
import matplotlib.pyplot as plt
import numpy as np
```

Se incluye la imagen que se va a utilizar

```python
img = cv2.imread('../resources/ex4.tif')
```

```python
plt.imshow(img)
plt.show()
```

```python
def extract_borders(image, border_width):
    """ Extracts the borders of an image with the specified width.
        @param image Image to extract borders
        @param border_width Size of the border
        @return Image iamge with only borders.
    """
    B = np.ones((border_width, border_width), np.uint8)  #Kernerl
    img_eroded = cv2.erode(image, B, iterations=1) 
    plt.imshow(img_eroded)
    plt.show()
    return image - img_eroded
```

```python
img_borders = extract_borders(img, 3)
plt.imshow(img_borders)
plt.show()
```

La extracción de bordes se realiza mediante primero erosionar la imagén original por un kernel cuadrado, esto justamente lo que hace es "borrarle" los bordes a la imagen original. Luego se hace la resta entre la imagen original y la imagen erosionada para obtener una imágen que contiene únicamente los bordes.
